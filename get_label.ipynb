```python
import nibabel as nib
import numpy as np

path = "./41_training_label/patient0039_gt.nii.gz"
nii = nib.load(path)
lab = nii.get_fdata()

print("unique:", np.unique(lab))

# 抓幾個 slice，可視化看看 1 和 2 長怎樣
z = lab.shape[2] // 2
slice_1 = (lab[..., z] == 1).astype(np.uint8)
slice_2 = (lab[..., z] == 2).astype(np.uint8)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,4))

plt.subplot(1, 2, 1)
plt.title("Label == 1 (check this)")
plt.imshow(slice_1, cmap="gray")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Label == 2 (check this)")
plt.imshow(slice_2, cmap="gray")
plt.axis("off")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 你原本的
z = lab.shape[2] // 2  # 取中間 slice

slice_img = lab[..., z]

# 分別取三個 class
slice_0 = (slice_img == 0).astype(np.uint8)
slice_1 = (slice_img == 1).astype(np.uint8)
slice_2 = (slice_img == 2).astype(np.uint8)

plt.figure(figsize=(15, 5))

plt.subplot(1, 4, 1)
plt.title("Raw Label Slice")
plt.imshow(slice_img, cmap='gray')
plt.axis("off")

plt.subplot(1, 4, 2)
plt.title("Label == 0 (Background)")
plt.imshow(slice_0, cmap='gray')
plt.axis("off")

plt.subplot(1, 4, 3)
plt.title("Label == 1 (Myocardium?)")
plt.imshow(slice_1, cmap='gray')
plt.axis("off")

plt.subplot(1, 4, 4)
plt.title("Label == 2 (?)")
plt.imshow(slice_2, cmap='gray')
plt.axis("off")

plt.show()
```
